```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.utils import to_categorical
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('/content/Breast_cancer_dataset.csv')
```

```python
convert = lambda x: 1 if x=='M' else 0
```

```python
df['target'] = df['diagnosis'].apply(convert)
```

```python
df.drop(columns='diagnosis',inplace=True,index=1)
```

```python
df.head()
```

```python
X = df.drop(columns='target')
y = df['target']
```

```python
X.head()
```

```python
y = to_categorical(y,num_classes=2)
```

```python
# train test split
X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.15,random_state=42)
X_train,X_test,y_train,y_test  = train_test_split(X_train,y_train,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_train_re,y_train_res = sm.fit_resample(X_train,y_train)
```

```python
print(X_train.shape,
X_val.shape,
X_test.shape)
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_val = scalar.transform(X_val)
X_test = scalar.transform(X_test)
```

```python
import keras
from keras.layers import Dropout
reg2 = keras.regularizers.L2(l2=0.01)
```

```python
inputs = Input(shape=(X_train.shape[1],))
layer1 = Dense(128,activation='relu',kernel_regularizer=reg2)(inputs)
dropout1 = Dropout(rate=0.2)(layer1)
layer2 = Dense(64,activation='relu',kernel_regularizer=reg2)(dropout1)
dropout2 = Dropout(rate=0.2)(layer2)
layer3 = Dense(32,activation='relu',kernel_regularizer=reg2)(dropout2)
layer4 = Dense(16,activation='relu',kernel_regularizer=reg2)(layer3)
layer5 = Dense(8,activation='relu')(layer4)
outputs = Dense(2,activation='sigmoid')(layer5)
```

```python
model = Model(inputs = inputs,outputs=outputs)
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=30,validation_data=[X_val,y_val])
```

```python
model.evaluate(X_test,y_test)
```

```python
model.evaluate(X_val,y_val)
```
